```python
import pandas as pd
```

```python
df = pd.read_csv("./data/구석구석_서울크롤링_4057.csv")
```

```python
# 컬럼 중 모든 데이터가 null인 컬럼
null_col = []

for col in df.columns :
    if df.loc[:, col].isnull().all() :
        print(col, 'is null')
        null_col.append(col)
```

```python
# 재확인
print("데이터 총 개수 :", len(df))
for col in null_col :
    print(col,'<< null 개수 :', df.loc[:, col].isnull().sum())
    # 개수가 같으면 drop
    if len(df) == df.loc[:, col].isnull().sum() :
        df.drop(columns=col, inplace=True)
```

#### 자연어 처리에 필요한 컬럼들은 '장소명', '한줄설명', '상세설명', '리뷰'

```python
import glob
new_image_df = pd.DataFrame()

for path in glob.glob("./data/image_src/*") :
    if path == './data/image_src/통합_구석구석_서울_이미지_4058.csv' :
        break
    
    image_df = pd.read_csv(path, index_col='Unnamed: 0')
    image_df.columns = ['사진주소', '장소명']

    new_image_df = pd.concat([new_image_df, image_df])

# new_image_df.to_csv("./data/image_src/구석구석_서울_이미지_4058.csv", index=None)
```

```python
# image_df와 df 병함
df = pd.merge(df, new_image_df, how='inner', left_on="장소명", right_on="장소명")
```

```python
print_arr = []
for index in range(len(df)) :
    df_dict = df.loc[index, :].dropna().to_dict()

    temp_arr = []
    for key, value in zip(df_dict.keys(), df_dict.values()) :
        temp_arr.append(f"{key} : {value}")
    print_arr.append(temp_arr)
print_arr
```
